Deberas crear tu propia funci칩n para calcular el error de reconstrucci칩n, que viene definido por:

$$SSE =  \sum_{n}^{i=1}  \begin{Vmatrix}x_{i} -  \widehat{x}_i \end{Vmatrix} ^2 $$

Donde:

- $x_i$ son los valores de la matriz original X
- $\widehat{x}_i$ son los valores de la matriz reconstruida

```python
### TODO: Funci칩n para calcular el error de reconstrucci칩n
### TU RESPUESTA ABAJO

import numpy as np

def sse_score(X, X_hat):
    return np.sum((X - X_hat)**2)

X = np.array([[1, 2], [3, 4]])
X_hat = np.array([[1.01, 1.75], [2.81, 3.99]])
sse = sse_score(X, X_hat) # -> 0.09879

print(sse)
```

Una vez que ya tenemos la funci칩n `sse` hecha, podemos pasar a construir la funci칩n que ejecutar치 `SVM`.

```python
### TODO: Funci칩n para ejecutar SVM
### Tiene como entrada una matriz X
### Devuelve U, s, Vt

### Hint: S debe ser una matriz diagonal
### TU RESPUESTA ABAJO

def svm(X):
    U, s, Vt = np.linalg.svd(X, full_matrices = False)
    S = np.diag(s)
    return U, S, Vt


# X = np.array([[1, 2], [3, 4]])
U, S, Vt = svm(X)


# -> U = array([[-0.40455358, -0.9145143 ],
# [-0.9145143 , 0.40455358]])
# -> S = array([[5.4649857 , 0. ],
# [0. , 0.36596619]])
# -> Vt = array([[-0.57604844, -0.81741556],
# [ 0.81741556, -0.57604844]])
```

Como hemos visto en clase, las matrices obtenidas a partir de `SVM` nos sirven para reconstruir la matriz original `X`. Para ello, construye una funci칩n que permita reconstruir la matriz original `X` a partir de `U, s, Vt`.

```python
### TODO: Funci칩n para reconstruir la matriz original a partir de U, s, Vt
### Tiene como entrada U, s, Vt
### Devuelve X_hat
### TU RESPUESTA ABAJO


def reconstruction(U, S, Vt):
    return U.dot(S.dot(Vt))

X_hat = reconstruction(U, S, Vt)


# X_hat -> array([[0.99999999, 1.99999998],
# [3.00000003, 4.00000001]])

print(X_hat)
```

```python
### TODO: Funci칩n que recibe una imagen A y devuelve la imagen comprimida
### Tiene como entrada A y el n칰mero de componentes para realizar la reducci칩n de dimensionalidad
### Devuelve la imagen comprimidad, el error de reconstrucci칩n y el ratio de compresi칩n

### Hint: Usa las funciones anteriormente construidas
### TU RESPUESTA ABAJO


def image_compression(A, n_components): # Corregida. Original mal implementada. Antes se utilizaba un bucle que no era necesario
    U, S, Vt = svm(A) # Descomponemos en tres matrices la imagen A. las matrices unitarias U y Vt y la matriz diagonal S

    U_reduc = U[:, :n_components] # Nos quedamos con las primeras n_components columnas de U
    S_reduc = S[:n_components, :n_components] # Las primeras n_components filas y columnas de S
    Vt_reduc = Vt[:n_components, :] #Las primeras n_components filas de Vt
    # Aqu칤 ya nos hemos quedado con la informaci칩n principal de la imagen reducida (comprimida)

    A_hat_reconst = reconstruction(U_reduc, S_reduc, Vt_reduc) # Ahora utilizamos la funci칩n reconstruction para reconstruir la
                      # matriz de la imagen a partir de las matrices reducidas
    SSE = sse_score(A, A_hat_reconst) # Calculamos el error de reconstrucci칩n comparando la matriz original y la reconstruida
    comp_ratio = (A.shape[1]*n_components + n_components + n_components*A.shape[0]) / (A.shape[0]*A.shape[1])
                      # Ahora calculamos en comp_ratio el ratio de compresi칩n

    return A_hat_reconst, SSE, comp_ratio

A_hat, sse, comp_ratio = image_compression(A, n_components=50)
print(A_hat)
print(sse)
print(comp_ratio)
```

Ahora, implementar치s la ecuaci칩n $w_{LS}$:

<center>$w_{LS} = (X^T X)^{1}X^T y,$</center>

```python
### TODO: Funci칩n para resolver la ecuaci칩n wLS
### Toma como argumentos dos matrices, una para X y otra para y
### Asumimos que las matrices tienen las dimensiones correctas

### Paso 1: Asegurate que n > d.
### Es decir, que el n칰mero de observaciones es mayor que el n칰mero de dimensiones.
### O lo que es lo mismo, que el n칰mero de filas de cada matriz sea mayor que el n칰mero de columnas
### Si no es as칤, debes transponer las matrices

### Paso 2: Debes a침adir a la matriz X un vector columna del tama침o (n x 1)

### Paso 3: Usa la ecuaci칩n de arriba para obtener wLS

### TU RESPUESTA ABAJO


def least_squares_weights(input_x, target_y):
    n_x, d_x = input_x.shape
    n_y, d_y = target_y.shape
    if not(n_x > d_x):
        input_x = input_x.T
        n_x, d_x = input_x.shape
    if not (n_y > d_y):
        target_y = target_y.T

    A = np.ones((n_x, 1))
    XA = np.concatenate((A, input_x), axis = 1)

    XA_t = np.transpose(XA)
    prod = np.dot(XA_t, XA)
    inv_prod_XA_t = np.linalg.inv(prod)

    XA_t_Y = np.dot(XA_t, target_y)

    wls = np.dot(inv_prod_XA_t, XA_t_Y)

    return wls


training_y = np.array([[208500, 181500, 223500,
                                140000, 250000, 143000,
                                307000, 200000, 129900,
                                118000]])
training_x = np.array([[1710, 1262, 1786,
                        1717, 2198, 1362,
                        1694, 2090, 1774,
                        1077],
                       [2003, 1976, 2001,
                        1915, 2000, 1993,
                        2004, 1973, 1931,
                        1939]])
weights = least_squares_weights(training_x, training_y)

print(weights)  #--> np.array([[-2.29223802e+06],
                #              [ 5.92536529e+01],
                #              [ 1.20780450e+03]])

print(weights[1][0])


    #Asumimos:
    #    -- target_y es un vector con el mismo n칰mero de observaciones que input_x
```

## 4. Linear Regression - Gradient Descent

En este ejercicio resolveras el mismo problema anterior pero usando **Gradient Descent**

**Objetivos**:
- Asegurar los fundamentos matem치ticos detr치s del Gradient Descent.

**Problema**: Usando datos sobre el precio de la vivienda, intentaremos predecir el precio de una casa en base a la superficie habitable con un modelo de regresi칩n.

**Datos:** [Kaggle's House Prices Dataset](https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data)

**Repaso:**

$$ RSS(w) = \sum_{n=1}^{N}[y_n-f(x_n)]^2 =  \sum_{n=1}^{N}[y_n- (w_0 + \sum_{d=1}^{D}w_dx_{nd}) ]^2 .$$

Loss function:

$$ RSS(w) = \frac{1}{2}\sum_{n=1}^{N}[y_n-f(x_n)]^2$$

Y lo que queremos es minimizar esta distancia, para que el modelo se acerque lo m치ximo posible a los valores verdaderos.

$$\nabla RSS(w) = X^T(Xw^t-y)$$

En resumen, el gradient descendiente para una regresi칩n lineal, se basa en resolver esta ecuaci칩n de forma iterativa:

$$w^{t+1} = w^t - \eta * \nabla RSS(w)$$

#### Leer Datos

```python
import pandas as pd
import numpy as np

# Leer datos
data = pd.read_csv('train.csv')

# Extraer dichas columnas
newData = data[['GrLivArea','SalePrice']]
print(newData.head())

# Contruir x - y
x = newData['GrLivArea']
y = newData['SalePrice']

# Standarizar los datos
x = (x - x.mean()) / x.std()
x = np.c_[np.ones(x.shape[0]), x]

print("Shape of X: ", x.shape)
print("Shape of y:", y.shape)
```

#### Gradient Descent

```python
### TODO: Funci칩n para encontrar los valores w usando Gradient Descent
### Toma como argumentos: X, y, w, n_iterations, eta
### Completa la funci칩n a침adiendo la loss funci칩n y la updating rule
### TU RESPUESTA ABAJO

def gradient_descent(x, y, w, iterations, eta): # Corregida. Mal implementada la primera vez
    """Gradient descent

    Argumentos:
        x -- Matriz con los datos de entrenamiento
        y -- Vector con los datos de salida
        w -- Vector aleatoriamente inicializado
        iterations -- N칰mero de iteraciones
        eta -- Learning Rate

    Ejemplo:
        import numpy as np

        # Learning rate
        eta = 0.01

        # N칰mero de iteraciones
        iterations = 2000 #No. of iterations

        # Seed para inicializar w
        np.random.seed(123)
        w0 = np.random.rand(2)

        training_y = np.array([208500, 181500, 223500,
                                140000, 250000])
        training_x = np.array([[ 1.        ,  0.37020659],
                               [ 1.        , -0.48234664],
                               [ 1.        ,  0.51483616],
                               [ 1.        ,  0.38352774],
                               [ 1.        ,  1.29888065]])

        weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)

        print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
    """
    past_loss = []
    past_w = [w]
    n = y.size
    for i in range(iterations):
        prediction = np.dot(x, w)
        error = prediction - y

        # Calculamos el loss
        loss = np.mean(error ** 2)
        past_loss.append(loss)

        # Calculamos el gradiente
        gradient = np.dot(x.T, error)

        # Actualizaci칩n de pesos
        w -= eta * gradient

        past_w.append(w)

    return past_w, past_loss

import numpy as np
# Learning rate
eta = 0.01

# N칰mero de iteraciones
iterations = 2000 #No. of iterations

# Seed para inicializar w
np.random.seed(123)
w0 = np.random.rand(2)

training_y = np.array([208500, 181500, 223500,
                        140000, 250000])
training_x = np.array([[ 1.        ,  0.37020659],
                        [ 1.        , -0.48234664],
                        [ 1.        ,  0.51483616],
                        [ 1.        ,  0.38352774],
                        [ 1.        ,  1.29888065]])

weights, loss = gradient_descent(training_x, training_y, w0, iterations, eta)

print(weights[-1])  #--> np.array([183845.82320222  40415.66453324])
```

Una vez que tenemos una lista de conjuntos por cada par de p치ginas web, podemos convertir el texto de la p치gina web a un vector.

```python
def set2vector(tokens_web1, tokens_web_2, set_web1, set_web2):
    """
    Funci칩n para convertir un conjunto a vector

    Argumentos:
        tokens_web1 -- Contenido tokenizado de p치gina web 1
        tokens_web_2 -- Contenido tokenizado de p치gina web 2
        set_web1 -- Conjunto de palabras de la p치gina web 1
        set_web2 -- Conjunto de palabras de la p치gina web 2

    Ejemplo:
        tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
        tokens_web_2 = ["hola", "me", "llamo"]
        set_web1 = {"hola", "que", "tal"}
        set_web2 = {"hola", "me", "llamo"}
        union_ls = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)

    Requerimientos:
        Depende de la funci칩n `union()`
    """

    # Unimos los conjuntos
    join_set = union([(set_web1, set_web2)])[0]

    web1_array = []
    web2_array = []

    for word in join_set:
        if word in tokens_web1:
            web1_array.append(1)
        else:
            web1_array.append(0)
        if word in tokens_web_2:
            web2_array.append(1)
        else:
            web2_array.append(0)

    return web1_array, web2_array

tokens_web1 = ["hola", "que", "tal", "soy", "Alex"]
tokens_web_2 = ["hola", "me", "llamo"]
set_web1 = {"hola", "que", "tal", "soy", "Alex"}
set_web2 = {"hola", "me", "llamo"}
web1_array, web2_array = set2vector(tokens_web1, tokens_web_2, set_web1, set_web2)
```

### Cosine Similarity

Por 칰ltimo, ya podemos implementar la ecuaci칩n: $\cos \theta = \frac{\overrightarrow{u} \cdot \overrightarrow{v}}{|\overrightarrow{u}||\overrightarrow{v}|}$

```python
### TODO: Escribe una funci칩n que recibe dos vectores, u y v
### Y devuelva la similaridad entre ambos vectores
###
### Paso 1: Si u y v son listas -> Convertirlo a arrays
###
### Paso 2: Calcula la similaridad entre ambos vectores
### TU RESPUESTA ABAJO

import numpy as np

def cosine_similarity(u, v):
    """Calcula la similaridad entre dos vectores

    Argumentos:
        u -- Vector 1
        v -- Vector 2

    Ejemplo:
        u = np.array([1, 2, 3])
        v = np.array([3, 2, 1])

        similarity = cosine_similarity(u, v)  # -> 0.71428
    """
```

```python
def websites_sim(url_ls):
    """Funci칩n para calcular la similaridad entre p치ginas web

    Argumentos:
        url_ls -- Listas de p치ginas web

    Ejemplo:
        url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris',
        'https://es.wikipedia.org/wiki/Canis_lupus',
        'https://es.wikipedia.org/wiki/Felis_silvestris_catus']

        similarity_ls = websites_sim(url_ls)
    """

    url2content = get_content(url_ls)

    # Creamos un diccionario donde cada url tendr치 su contenido tokenizado y su conjunto
    url_dict = {}
    for url, content in url2content.items():
        toks, sets = tokenizer(content)
        url_dict[url] = {'tokens': toks,
                        'unique_tokens': sets}

    # Calculamos todas las combinaciones posibles de las direcciones de las p치ginas web
    comb_ls = combinations(url_ls)

    # Usando comb_ls y la funci칩n `set2vector()` convertimos cada p치gina web a vectores
    print("Similaridad: ")
    for el in comb_ls:
        # Obtenemos los tokens y el conjunto para cada p치gina web
        token_1 = url_dict[el[0]]['tokens']
        token_2 = url_dict[el[1]]['tokens']
        set_1 = url_dict[el[0]]['unique_tokens']
        set_2 = url_dict[el[1]]['unique_tokens']
        array_web1, array_web2 = set2vector(token_1, token_2, set_1, set_2)
        similarity = cosine_similarity(array_web1, array_web2)

        print("{} vs {} - {}".format(el[0], el[1], round(similarity, 3)))


url_ls = ['https://es.wikipedia.org/wiki/Canis_lupus_familiaris',
'https://es.wikipedia.org/wiki/Canis_lupus',
'https://es.wikipedia.org/wiki/Felis_silvestris_catus']

similarity_ls = websites_sim(url_ls)
```
